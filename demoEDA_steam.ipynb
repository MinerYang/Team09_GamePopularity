```python
import pandas as pd
import numpy as np
data = pd.read_csv("/Users/mineryang/Desktop/steam-store-games/steam.csv")
data.shape
```

## 原始数据

```python
df = data
df.head()
```

## 处理 platform,genres,categories

```python
platform_set = set()
for i in df.platforms.str.split(';'):
    platform_set.update(i)
platform_set

platform_sets = df.platforms.str.split(';').apply(set)
d = dict()
d['windows'] = platform_sets.apply(lambda row: 'windows' in row)
d['linux'] = platform_sets.apply(lambda row: 'linux' in row)
d['mac'] = platform_sets.apply(lambda row: 'mac' in row)
df = df.assign(**d)
```

```python
df.head()
```

### 处理genre

```python
genres_set = set()
for i in df.genres.str.split(';'):
    genres_set.update(i)
genres_set

d = dict()
genre_sets = df.genres.str.split(';').apply(set)
for genre in genres_set:
    d[genre] = genre_sets.apply(lambda row: genre in row)
df = df.assign(**d)
```

```python
df.head()
```

```python
df = df.drop(columns=['Utilities', 'Design & Illustration','Animation & Modeling','Education','Video Production','Software Training',
                'Audio Production','Web Publishing','Game Development','Photo Editing','Accounting','Tutorial','Documentary'])
```

### 处理catagories

```python
categories_set = set()
for i in df.categories.str.split(';'):
    categories_set.update(i)
categories_set

d = dict()
category_set = df.categories.str.split(';').apply(set)
for category in categories_set:
    d[category] = category_set.apply(lambda row: category in row)
df = df.assign(**d)
```

```python
df = df.drop(columns=['VR Support','Commentary available','Valve Anti-Cheat enabled','Steam Turn Notifications','Includes Source SDK',
                'SteamVR Collectibles','Mods','Mods (require HL2)'])
```

```python
mask = df1.owners.isin(['10000000-20000000', '20000000-50000000', '50000000-100000000', '100000000-200000000'])
df1.loc[mask, 'owners'] = 10

df1.loc[df1.owners == '5000000-10000000', 'owners'] = 9
df1.loc[df1.owners == '2000000-5000000', 'owners'] = 8
df1.loc[df1.owners == '1000000-2000000', 'owners'] = 7
df1.loc[df1.owners == '500000-1000000', 'owners'] = 6
df1.loc[df1.owners == '200000-500000', 'owners'] = 5
df1.loc[df1.owners == '100000-200000', 'owners'] = 4
df1.loc[df1.owners == '50000-100000', 'owners'] = 3
df1.loc[df1.owners == '20000-50000', 'owners'] = 2
df1.loc[df1.owners == '0-20000', 'owners'] = 1
```
